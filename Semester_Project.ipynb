```python
import cv2  # For computer vision tasks.
import face_recognition  # For face recognition tasks.
from matplotlib import pyplot as plt  # For image visualization.
```

```python
# Load the image using OpenCV
img = cv2.imread("rizwan.jpeg")

# Convert the image from BGR (Blue, Green, Red) to RGB (Red, Green, Blue) format
rgb_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Display the image using matplotlib
plt.imshow(rgb_img)
plt.axis('off')  # Turn off axis numbers and ticks
plt.show()

# Compute the facial encodings of the detected faces in the image using the face-recognition library
face_encodings = face_recognition.face_encodings(rgb_img)

# Check if any faces are detected in the image
if len(face_encodings) > 0:
    # If faces are detected, print the number of faces detected in the image
    print("Number of faces detected in the image:", len(face_encodings))

    # Extract the facial encoding of the first face detected
    img_encoding = face_encodings[0]
else:
    # If no faces are detected, print a message
    print("No faces are detected in the image.")
```
